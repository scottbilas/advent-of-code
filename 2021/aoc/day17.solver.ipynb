The probe's x,y position starts at 0,0. Then, it will follow some trajectory by moving in steps. On each step, these changes occur in the following order:

- The probe's x position increases by its x velocity.
- The probe's y position increases by its y velocity.
- Due to drag, the probe's x velocity changes by 1 toward the value 0; that is, it decreases by 1 if it is greater than 0, increases by 1 if it is less than 0, or does not change if it is already 0.
- Due to gravity, the probe's y velocity decreases by 1.

For the probe to successfully make it into the trench, the probe must be on some trajectory that causes it to be within a target area after any step. The submarine computer has already calculated this target area (your puzzle input). For example:

```python
def parse(text):
    return [int(v) for v in re.findall(r'-?\d+', text)]

def solve1(text):
    ymin, ymax = parse(text)[2:]

    def test(vel):
        my, y, dy = 0, 0, vel
        while y >= ymin:
            ly = y
            y += dy
            my = max(my, y)
            dy -= 1

        return my if ly >= ymin and ly <= ymax else None

    my = 0
    for i in range(1, 1000):
        ty = test(i)
        if ty:
            my = max(my, ty)

    return my

check(solve1(_sample), 45)
check1(solve1(_input))
```

```python
from collections import defaultdict

def solve2(text):
    xmin, xmax, ymin, ymax = parse(text)

    xhits = defaultdict(lambda: []) # {<steps>: [ivx, ...]}

    for ivx in range(xmax+1):
        steps, x, vx = 0, 0, ivx
        while vx > 0:
            steps += 1
            x += vx
            vx = max(vx-1, 0)
            if x>=xmin and x<=xmax:
                xhits[steps].append(ivx)
                if vx == 0:
                    for _ in range(1000):
                        steps += 1
                        xhits[steps].append(ivx)

    hits = set()

    ivy = ymin
    while True:
        steps, y, vy = 0, 0, ivy

        while True:
            steps += 1
            y += vy
            vy -= 1
            if y>=ymin and y<=ymax:
                for ivx in xhits[steps]:
                    hits.add((ivx, ivy))
            if y < ymin:
                break
        ivy += 1
        if ivy > 1000:
            break

    return len(hits)

#check(solve2(_sample), 112)
check2(solve2(_input))   # 3446 and 3620 too low, 3786 too high
```
