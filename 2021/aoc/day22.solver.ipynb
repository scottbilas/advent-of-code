```python
def parse(text):
    onoff, nums = re.findall('on|off', text), [int(v) for v in re.findall('-?\d+', text)]
    return [(onoff[i//6] == 'on', tuple(nums[i:i+6])) for i in range(0, len(nums), 6)]

def solve(cuboids):

    # TODO: 8s is way too slow, so set up a simple LUT or hashtable for finding candidate intersections

    def intersect(c0, c1):
        x0, x1 = max(c0[0], c1[0]), min(c0[1], c1[1])
        if x1 < x0: return None
        y0, y1 = max(c0[2], c1[2]), min(c0[3], c1[3])
        if y1 < y0: return None
        z0, z1 = max(c0[4], c1[4]), min(c0[5], c1[5])
        if z1 < z0: return None
        return x0, x1, y0, y1, z0, z1

    out = []
    for srcOn, srcBox in cuboids:
        oldLen = len(out)
        if srcOn:
            out.append((srcOn, srcBox))
        for i in range(oldLen):
            testOn, testBox = out[i]
            if overlap := intersect(srcBox, testBox):
                out.append((not testOn, overlap))

    def volume(cuboid):
        on, coords = cuboid[0], cuboid[1]
        vol = (coords[1]-coords[0]+1) * (coords[3]-coords[2]+1) * (coords[5]-coords[4]+1)
        return vol if on else -vol

    return sum(map(volume, out))

sampleData1 = parse(_sample1)
sampleResults1 = (
    solve(sampleData1[:1]),
    solve(sampleData1[:2]),
    solve(sampleData1[:3]),
    solve(sampleData1))
check(sampleResults1[0], 27)
check(sampleResults1[1], sampleResults1[0]+19)
check(sampleResults1[2], sampleResults1[1]-8)
check(sampleResults1[3], sampleResults1[2]+1)
check(sampleResults1[3], 39)

def solve1(text):
    return solve([v for v in parse(text) if max(map(abs, v[1])) <= 50])

check(solve1(_sample2), 590784)
check1(solve1(_input))
```
